```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
```

```python
# generate some data
x = np.random.normal(3.0, 1.0, 1000)
y = 100 - (x + np.random.normal(0, .1, 1000)) * 3
```

```python
# make scatter plot
plt.scatter(x,y)
```

```python
# retrieve linear regression values
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
```

```python
# calculate r^2 value
r2 = r_value**2
r2
```

```python
# determine line of best fit
def predict(X):
    return slope * X + intercept

fitLine = predict(x)

# plot data with line of best fit
plt.clf()
plt.scatter(x, y, label='Data')
plt.plot(x, fitLine, 'r-', label='Best Fit')
plt.legend()
plt.show()
```
